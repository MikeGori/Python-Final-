### Part 1 - A - Visualizations of infection spread
Write a Python function called `simulate_covid_and_lockdown()`. This function receives 5 parameters (inputs):
1. The (initial) number of people infected.
2. The R rate (i.e., how many people a person with COVID infects per-day).
3. Number of days to simulate.
4. The day # in-which a lock-down is initiated (must be < # of days to simulate).
5. The R rate after lock-down.

The function will 'simulate' how the # infected people rises, then drops after lock-down. The function visualizes this simulation (you can use the above code example). The function **does not** need to return any values.

*NOTE*: Use a FOR loop to simulate, not a determinstic equation.

```python
import matplotlib.pyplot as plt

def simulate_covid_and_lockdown(ininfected, rpre, days, lock, rpost):
    if lock > days:
         print('The days simulated must be larger than the days until lockdown')
    else:
        
        '''
        Empty list to plot from
        '''
        
        inf = []
        
        '''
        The first value of the list is the initial number of infected people
        '''
        
        inf.append(ininfected)
        
        '''
        Looping on the range of days before the lockdown and adding each calculated number to the inf list
        '''
        
        for i in range(lock):
            prog = (inf[-1] * rpre) 
            inf.append(round(prog))
            
            ''' 
            values post the lockdown
            '''
            
        for i in range(days - lock):
            prog2 = (inf[-1] * rpost)
            inf.append(round(prog2))
            
            '''
            breaking the loop when the values hit 0
            '''
            
            if inf[-1] <= int(0):
                break
            
                                    
    
    inf_len= len(inf)
    days_list = list(range(inf_len))
                            
    plt.plot(days_list, inf, '-o')
    plt.title('Number of infected by day, based on constant {0}'.format(rpre))
    plt.xlabel('# Days since {0} persons infected'.format(ininfected))
    plt.ylabel('# of people infected')
    plt.show()
               
    
    
    
```

The function also prints: def simulate_to_end_lockdown(ininfected, rpost):
    
   

```python
def simulate_to_end_lockdown(ininfected, rpost):
    
    '''
    The R value has to be lower than 1
    '''
    if rpost > 1:
        print('The R value must be lower than 1')
    else:
        
        '''
        Creating an empty list and appending the first value of infected people
        '''
        inf = []
        inf.append(ininfected)
        
        '''
        Making a while loop to check whether the amount of infected is lower tha 100
        and appending each result to to the list to count the loops, each loop equals a day
        '''
        while round(inf[-1] * rpost) > 100:
            days = inf[-1] - (inf[-1] * rpost)
            inf.append(days)
            
        print("Given your input parameters, it will take {0} days until there are less than 100 people infected.".format(len(inf)))
            
```

Notes:
- Ignore upper and lower case letters (treat them as the same)
- Ignore (you can remove) punctuation and symbols (characters such as "- . ! ,"). So Sam-I-Am = sam i am.

```python
import re


def get_unique_words(data):
    """
    Returns a dictionary: {word} : {apperances}
    """
    seen_words = {}
    words = data.split(' ')
    for word in words:
        current_value = seen_words.setdefault(word, 0)
        seen_words[word] = current_value + 1

    return seen_words
    

def strip_data(data):
    data = data.lower()
    data = data.replace('\r\n', ' ')
    data = re.sub('[-!,"]+', ' ', data)
    data = data.replace(".", ' ')
    data = data.replace("?", ' ')
    data = data.replace("\\", ' ')
    return data

def get_file_content(file_path):
    """
    Gets a path of a file and returns all the 
    data inside it
    """
    data = ""
    with open(file_path, 'rb') as f:
        data = f.read()

    return data.decode('utf-8')

def get_unique_of_books(first_path, second_path):
    """
    Gets two paths and returns two lists of unique words
    """
    data_1 = get_file_content(first_path)
    data_1 = strip_data(data_1)
    data_2 = get_file_content(second_path)
    data_2 = strip_data(data_2)

    unique_1 = get_unique_words(data_1)
    unique_2 = get_unique_words(data_2)

    return unique_1, unique_2

def compare_unique_words(first_path, second_path):
    """
    Answers the first qustion
    """
    unique_dict_1, unique_dict_2 = get_unique_of_books(first_path, second_path)
    unique_1 = unique_dict_1.keys()
    unique_2 = unique_dict_2.keys()

    has_most_unique_words = first_path if len(unique_1) > len(unique_2) else second_path
    print ("Most unique words: {0}".format(has_most_unique_words))
    print ("Number unique words first book: {0}".format(len(unique_1)))
    print ("Number unique words second book: {0}".format(len(unique_2))) 	

def how_many_times(first_path, second_path):
    """
    Answers the Second question
    """
    unique_1, unique_2 = get_unique_of_books(first_path, second_path)
    print("First book unique words apperances: {0}".format(unique_1))
    print("Second book unique words apperances: {0}".format(unique_2))
```

## Part 4 - **Hat game**
In this part, you will write a function called `hat_game(N)` that simulates the outcome of the hat game of *N* (*N* > 10) people.

In the hat game:
- *N* persons are sitted in a circle. Each person brings with him a hat and places it on their lap. Each person has a number, from #1 to #*N*.
- You, the player, go around the circle *N* times, every time starting at Person #1 and finishing at Person #*N*.
- The **1st** time around you stop at **every person** (e.g., Person 1, person 2, person 3 ... person). The **2nd** time around you stop at every **second** person (Person 2, Person 4, person 6 ...). The **3rd** time around you stop at every **third** person (Person 3, person 6 ...) and so on *N* times around.
- Each person you stop at, you either place the hat on their head, or remove back to their lap. So:
    - if they *dont* have a hat on, you put a hat on their head.
    - If they *do* have a hat on, you take it off.

The function receives one parameter (N) and returns a single `list` variable. The returned variable contains a list of person's numbers with **hats on** after you finished going around *N* times. E.g., the output might look like:

[2, 5, 78, 90] **(This is just an example of what the return might look like, the numbers are INCORRECT)**

```python
def filp_value(boolean_value):
    return not boolean_value

def hat_game(n):
    hats_status = {}

    # init dictionary
    for i in range(1, n + 1):
        hats_status[i] = False

    for i in range(1, n + 1):
        j = 1
        while j <= len(hats_status.keys()):
           
            hats_status[j] = filp_value(hats_status[j])
            j += i

    return [i for i in range(1, n + 1) if hats_status[i]]
```
